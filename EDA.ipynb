## Calculate Progression Length

## Visualize FVC Progression over Time

### Subtask:
Select a few representative patients and visualize their FVC measurements over time. This will help conceptualize different FVC decline curves (fast, slow, stable) and understand the heterogeneous progression patterns (Task T6).


**Reasoning**:
To identify representative patients with different FVC progression patterns (fast decline, slow decline, stable), I need to calculate the FVC decline rate (slope) for each patient. I will use a simple linear regression to determine the slope of FVC over 'Weeks' for each patient.



```python
from scipy.stats import linregress

def calculate_fvc_slope(patient_df):
    if len(patient_df) < 2:  # Need at least two points to calculate a slope
        return None
    slope, intercept, r_value, p_value, std_err = linregress(patient_df['Weeks'], patient_df['FVC'])
    return slope

# Calculate the slope of FVC over Weeks for each patient
fvc_slopes = df.groupby('Patient').apply(calculate_fvc_slope)

# Drop patients for whom a slope couldn't be calculated (e.g., only one measurement)
fvc_slopes = fvc_slopes.dropna()

print("Calculated FVC slopes for each patient:")
print(fvc_slopes.head())
```

**Reasoning**:
The previous step successfully calculated FVC slopes for each patient, despite a deprecation warning. Now, to fulfill the task of visualizing FVC progression for representative patients, I will select patients with fast decline, slow decline, and stable FVC based on these slopes, and then plot their FVC over time.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort slopes to identify representative patients
sorted_slopes = fvc_slopes.sort_values()

# Select representative patients:
# 1. Fast decline (most negative slope)
fast_decline_patient = sorted_slopes.index[0]
# 2. Slow decline (a moderately negative slope, e.g., around the 25th percentile)
slow_decline_patient = sorted_slopes.index[int(len(sorted_slopes) * 0.25)]
# 3. Stable FVC (slope closest to zero, could be slightly positive or negative)
stable_patient = sorted_slopes.index[(sorted_slopes - 0).abs().argsort()[0]]

# Filter data for selected patients
representative_patients_ids = [fast_decline_patient, slow_decline_patient, stable_patient]
representative_df = df[df['Patient'].isin(representative_patients_ids)].copy()

# Plot FVC progression over time for representative patients
plt.figure(figsize=(12, 7))
sns.lineplot(data=representative_df, x='Weeks', y='FVC', hue='Patient', marker='o')
plt.title('FVC Progression Over Time for Representative Patients')
plt.xlabel('Weeks')
plt.ylabel('FVC')
plt.grid(True)
plt.legend(title='Patient ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print(f"Selected representative patients for visualization:\nFast Decline: {fast_decline_patient} (Slope: {fvc_slopes[fast_decline_patient]:.2f})\nSlow Decline: {slow_decline_patient} (Slope: {fvc_slopes[slow_decline_patient]:.2f})\nStable FVC: {stable_patient} (Slope: {fvc_slopes[stable_patient]:.2f})")
```
